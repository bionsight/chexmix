```python
import networkx as nx
import pandas as pd
import chexmix.graph as graph
from chexmix.datasources.classyfire import query
```

```python
# input: list of smiles
queries = query([
    "CCS(=O)(=O)c1ccc(C(NC(=O)Cc2nccs2)C(=O)Nc2ccc(C(O)(C(F)(F)F)C(F)(F)F)cc2)nc1",
    "CC(C)(N)C(=O)N[C@@H]1CC[C@@]2(S(=O)(=O)c3ccc(F)cc3)c3ccc(C(F)(C(F)(F)F)C(F)(F)F)cc3CC[C@@H]12",
    "CCS(=O)(=O)c1ccc(CC(=O)Nc2ccc(CN3Cc4ccccc4C3)cc2)cc1",
    "CCS(=O)(=O)c1ccc(C(CC(=O)N(C)C)C(=O)Nc2ccc(C(O)(C(F)(F)F)C(F)(F)F)cc2)cc1"
])
```

```python
classyfire_graph=graph.ClassyFireGraph().from_classyfire_entities(queries)
```

```python
classyfire_graph.get_table()
```
