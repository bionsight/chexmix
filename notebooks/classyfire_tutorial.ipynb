```python
# prepare smiles
smiles = ['COC(=O)/C=C/C1=CC(=C(C=C1)O)O', 'COC(=O)/C=C/C1=CC=C(C=C1)O', 'CC(=CCC1=C(C=C(C2=C1O[C@@H]([C@H](C2=O)O)C3=CC=C(C=C3)O)O)O[C@H]4[C@@H]([C@H]([C@@H]([C@H](O4)CO)O)O)O)C', 'CC(C)(CCC1=C(C=C(C2=C1OC(=C(C2=O)O)C3=CC=C(C=C3)O)O)O[C@H]4[C@@H]([C@H]([C@@H]([C@H](O4)CO)O)O)O)O', 'C1=CC(=CC=C1C2=C(C(=O)C3=C(C=C(C=C3O2)O)O)O[C@H]4[C@@H]([C@H]([C@H]([C@H](O4)CO)O)O)O)O', 'C1=CC(=CC=C1C2=C(C(=O)C3=C(C=C(C=C3O2)O)O)O[C@H]4[C@@H]([C@H]([C@@H]([C@H](O4)CO)O)O)O)O', 'C1=CC(=C(C=C1C2=C(C(=O)C3=C(C=C(C=C3O2)O)O)O)O)O', 'CC(=C)[C@@H]1CC[C@]2([C@H]1[C@H]3CC[C@@H]4[C@]5(CCC(=O)C([C@@H]5CC[C@]4([C@@]3(CC2)C)C)(C)C)C)C', 'COC1=C(C=CC(=C1)/C=C/C(=O)OC)O', 'CC(C)(CCC1=C(C=C(C2=C1OC(=C(C2=O)O)C3=CC=C(C=C3)O)O)O[C@H]4[C@@H]([C@H]([C@@H]([C@H](O4)CO)O)O)O)O']
```

```python
# query to classyfire.com
q = query(smiles)
# get entities
classyfire_graph=ClassyFireGraph().from_classyfire_entities(q)
```
